```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("last_name_data/linear_train.txt")
test = pd.read_csv("last_name_data/linear_test.txt")
sample_submission = pd.read_csv("last_name_data/linear_ans_example.txt")
```

```python
train.head(10)
```

```python
def features(x, suffixes, letters):
    fs =  [len(x), 
           len(x.split())]
    for y in suffixes:
        n = len(y)
        if x[-n:] == y:
            fs.append(1)
        else:
            fs.append(0)
    for y in letters:
        if len(x) == len(y) and ''.join(sorted(list(x))) == y:
            fs.append(1)
        else:
            fs.append(0)
    return fs

def convert_to_features(X, suffixes, letters):
    return pd.DataFrame([features(x, suffixes, letters) for x in X])
```

```python
def all_suffixes(last_names):
    suffixes = set()
    for x in last_names:
        for i in range(len(x)):
            suffixes.add(x[-i])
    return suffixes
suffixes = all_suffixes(train['last name'])
```

```python
'abac'[-1:]
```

```python
def all_letters(last_names):
    letters = set()
    for x in last_names:
        letters.add(''.join(sorted(list(x))))
    return letters
letters = all_letters(train['last name'])
```

```python
X_train = convert_to_features(train['last name'], suffixes, letters)
y_train = train['y'] * 2 - 1
X_test = convert_to_features(test['last name'], suffixes, letters)
```

```python
X_train.head()
```

```python
from scipy import interp
from itertools import cycle

from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import StratifiedKFold
```

```python
def cross_val_roc_auc(X, y):
    cv = StratifiedKFold(n_splits=6)
    classifier = LogisticRegression()

    mean_tpr = 0.0
    mean_fpr = np.linspace(0, 1, 100)

    colors = cycle(['cyan', 'indigo', 'seagreen', 'yellow', 'blue', 'darkorange'])
    lw = 2

    i = 0
    plt.figure(figsize=(17, 10))
    for (train, test), color in zip(cv.split(X, y), colors):
        probas_ = classifier.fit(X.iloc[train], y.iloc[train]).predict_proba(X.iloc[test])
        fpr, tpr, thresholds = roc_curve(y[test], probas_[:, 1])
        mean_tpr += interp(mean_fpr, fpr, tpr)
        mean_tpr[0] = 0.0
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, lw=lw, color=color,
                 label='ROC fold %d (area = %0.2f)' % (i, roc_auc))

        i += 1
    plt.plot([0, 1], [0, 1], linestyle='--', lw=lw, color='k',
             label='Luck')

    mean_tpr /= cv.get_n_splits(X, y)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    plt.plot(mean_fpr, mean_tpr, color='g', linestyle='--',
             label='Mean ROC (area = %0.2f)' % mean_auc, lw=lw)

    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
```

```python
cross_val_roc_auc(X_train, y_train)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_test = model.predict(X_test)
```

```python
ans = pd.DataFrame()
ans['Id'] = list(range(len(y_test)))
ans['Answer'] = (y_test + 1) // 2
ans.head(5)
```

```python
ans.to_csv("garkavyy_last_name_logistic_regression.txt", sep=',', index=False)
```
